```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set a consistent style for plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6) # Default figure size for consistency
plt.rcParams['figure.dpi'] = 100 # Adjust for higher resolution plots
```

```python
# Load the dataset
df = pd.read_csv("C:\\Users\\GayathriHarshitha\\disney_plus\\disney_plus_titles.csv")

# Display the first few rows to understand the structure
print("First 5 rows of the dataset:")
print(df.head())

# Display general information to check data types and missing values
print("\nInformation about the dataset:")
print(df.info())

# Display descriptive statistics for numerical columns
print("\nDescriptive statistics of the dataset:")
print(df.describe())
```

```python
# --- 3. Data Cleaning and Preparation ---
print("\n--- Data Cleaning and Preparation ---")

# Convert 'date_added' to datetime objects and extract 'year_added'
# 'errors='coerce' will turn any parsing errors into NaT (Not a Time)
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year

# Handle missing values using direct assignment to avoid FutureWarning
# Fill missing 'country' with 'Unknown' as it's a categorical feature
df['country'] = df['country'].fillna('Unknown')

# Fill missing 'director' with 'Unknown' for content without a listed director
df['director'] = df['director'].fillna('Unknown')

# Parse 'duration' column:
# Create separate DataFrames for Movies and TV Shows for specific duration handling
movies_df = df[df['type'] == 'Movie'].copy()
tvshows_df = df[df['type'] == 'TV Show'].copy()

# For movies, extract minutes from 'duration' (e.g., '91 min' -> 91.0)
# Using regex=False for simple string replacement
movies_df['duration_minutes'] = movies_df['duration'].str.replace(' min', '', regex=False).astype(float)

# For TV shows, extract number of seasons (e.g., '1 Season' -> 1.0)
# Using regex=True to replace ' Season' or ' Seasons'
tvshows_df['duration_seasons'] = tvshows_df['duration'].str.replace(' Season|s', '', regex=True).astype(float)

print("Data Cleaning and Preparation Complete. Warnings regarding inplace=True should be resolved.")
print("Check for remaining missing values (should be mostly handled):")
print(df.isnull().sum()) # Shows remaining NaNs if any (e.g., in 'cast', 'rating', 'date_added' original NaNs)
```

```python
# --- 4. Key Calculations ---
print("\n--- Key Calculations ---")

# Calculate the average movie duration
average_movie_duration = movies_df['duration_minutes'].mean()
print(f"Average Movie Duration on Disney+: {average_movie_duration:.2f} minutes")

# You can add more calculations here as needed for your analysis:
# print(f"Number of unique directors: {df['director'].nunique()}")
# print(f"Number of unique countries: {df['country'].nunique()}")
# print(f"Most common rating: {df['rating'].mode()[0]}")
```

```python
plt.figure(figsize=(4,6)) # Slightly larger for better clarity

content_type_counts = df['type'].value_counts()
total_content = df.shape[0]

# Prepare labels including count and percentage for clarity
labels = [f'{label}: {count} ({count/total_content:.1%})'
          for label, count in content_type_counts.items()]

plt.pie(content_type_counts, labels=labels, autopct='', startangle=140,
        colors=['#66b3ff','#99ff99'], # Example colors for Movies/TV Shows
        wedgeprops={'edgecolor': 'black'}) # Add black edge for better separation

plt.title('Distribution of Content Types on Disney+ (Movies vs. TV Shows)', fontsize=16)
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a perfect circle.

# Add total content count as text for context
plt.text(0.5, 1.05, f"Total Content: {total_content} titles", fontsize=12,
         ha='center', va='bottom', transform=plt.gca().transAxes) # Centered above the pie chart

plt.tight_layout() # Adjust layout to prevent labels/title overlapping
plt.savefig('content_type_distribution_pie_chart.png') # Save the plot
plt.show() # Display the plot
```

```python
# --- Visualization 2: Number of Titles Added Per Year (Bar Chart - Refined) ---
# Get the count of titles added per year and sort by year
content_added_per_year = df['year_added'].value_counts().sort_index().dropna()

# Filter to years with actual content additions, typically starting from 2019 for Disney+
content_added_for_plot = content_added_per_year[content_added_per_year.index >= 2019]

# Convert the index (years) to integer type for display on the x-axis (removes .0)
content_added_for_plot.index = content_added_for_plot.index.astype(int)

plt.figure(figsize=(10, 6)) # Adjust figure size for better readability

# Use a bar plot to clearly show yearly volumes
sns.barplot(x=content_added_for_plot.index, y=content_added_for_plot.values, palette='Blues_d') # A sequential color palette

# Add numerical labels on top of the bars for exact values
for index, value in enumerate(content_added_for_plot.values):
    plt.text(index, value + 10, f'{int(value)}', ha='center', va='bottom', fontsize=10)

plt.title('Number of Titles Added to Disney+ Per Year (2019-2021)', fontsize=16)
plt.xlabel('Year Content Was Added', fontsize=12)
plt.ylabel('Number of Titles Added', fontsize=12)
plt.xticks(rotation=0) # Keep x-axis labels horizontal for clarity
plt.ylim(0, content_added_for_plot.values.max() * 1.15) # Set y-limit to give space for labels
plt.grid(axis='y', linestyle='--', alpha=0.7) # Only y-axis grid lines for cleaner look

plt.tight_layout()
plt.savefig('titles_added_per_year_bar_chart.png') # Save the plot
plt.show() # Display the plot
```

```python
# --- Visualization 3: Top 10 Directors by Number of Titles (Bar Chart) ---
plt.figure(figsize=(12, 6))

# Exclude 'Unknown' directors and get the top 10 most frequent directors
# errors='ignore' ensures it doesn't fail if 'Unknown' isn't in the list for some reason
top_directors = df['director'].value_counts().drop('Unknown', errors='ignore').head(10)

sns.barplot(x=top_directors.index, y=top_directors.values, palette='viridis') # Viridis is a perceptually uniform palette

plt.title('Top 10 Directors by Number of Titles on Disney+', fontsize=16)
plt.xlabel('Director', fontsize=12)
plt.ylabel('Number of Titles', fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.tight_layout()
plt.savefig('top_directors_bar_chart.png') # Save the plot
plt.show() # Display the plot
```

```python
# --- NEW VISUALIZATION 7: Distribution of Content Ratings (Count Plot) ---
plt.figure(figsize=(10, 6))

# Order the ratings if desired, e.g., for general maturity progression
# This is a subjective order and can be adjusted
rating_order = ['TV-Y', 'TV-Y7', 'TV-G', 'TV-PG', 'G', 'PG', 'TV-14', 'PG-13', 'Unknown', 'R']
# Filter out ratings not in your dataset or order them based on actual counts
actual_ratings = df['rating'].value_counts().index.tolist()
ordered_actual_ratings = [r for r in rating_order if r in actual_ratings]
# Add any ratings not in the predefined order to the end
for r in actual_ratings:
    if r not in ordered_actual_ratings:
        ordered_actual_ratings.append(r)


sns.countplot(x='rating', data=df, order=ordered_actual_ratings, palette='coolwarm')

plt.title('Distribution of Content Ratings on Disney+', fontsize=16)
plt.xlabel('Content Rating', fontsize=12)
plt.ylabel('Number of Titles', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('content_ratings_distribution_countplot.png')
plt.show()
```

```python

# Assuming df is your DataFrame, loaded and cleaned as per previous steps

# First, aggregate the data to count Movies and TV Shows per release year
yearly_counts = df.groupby(['release_year', 'type']).size().unstack(fill_value=0).reset_index()

# Display the first few rows of the aggregated data to understand its structure
print("Aggregated yearly counts for Movies and TV Shows:")
print(yearly_counts.head())

# Now, create the line plot
plt.figure(figsize=(15, 8)) # Make the figure larger for better readability

# Plot for Movies
sns.lineplot(data=yearly_counts, x='release_year', y='Movie', label='Movies', marker='o', linewidth=2)

# Plot for TV Shows
sns.lineplot(data=yearly_counts, x='release_year', y='TV Show', label='TV Shows', marker='o', linewidth=2)

# Set clear titles and labels
plt.title('Number of Movies and TV Shows Released Per Year', fontsize=18, fontweight='bold')
plt.xlabel('Release Year', fontsize=14)
plt.ylabel('Number of Titles Released', fontsize=14)

# Customize the legend
plt.legend(title='Content Type', loc='upper left', bbox_to_anchor=(1, 1), fontsize=12, title_fontsize='14')

# Add grid for easier reading
plt.grid(True, linestyle='--', alpha=0.6)

# Rotate x-axis labels for better readability if there are many years
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)

# Adjust layout to prevent labels from overlapping
plt.tight_layout()

# Adjust the right margin to make space for the legend
plt.subplots_adjust(right=0.88)

plt.show()
```
